```python
#putting feature variable to x
X=ad_multi[['TV','Radio','Newspaper']]

# Putting response variable to y
y=ad_multi['Sales']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,test_size=0.3,random_state=100)
```

```python
X_train
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LinearRegression    # from sklearn package we import linear_model module and using LinearRegression function
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)         # Intercepet Beta0
```

```python
coeff_df=pd.DataFrame(lm.coef_,X_test.columns,columns=['Coefficient'])
```

```python
coeff_df
```

```python
y_pred=lm.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score           
mse = mean_squared_error(y_test,y_pred)
r_squared = r2_score(y_test,y_pred)
```

```python
print('Mean_squared_Error:',mse)
print('r_square_value:',r_squared)
```
